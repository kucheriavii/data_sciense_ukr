```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("./train.csv")
```

```python
import seaborn as sns
```

```python
#Перетворити ознаку статі в цифру
data['Sex'] = np.where(data['Sex'] == 'male', 0, 1)
#Перетворити порти в цифрм
le = LabelEncoder()
data["Embarked"] = le.fit_transform(data['Embarked'])

data['Pizdatyi'] = data['Sex']/data['Pclass']
```

```python
sns.heatmap(data[["Survived", "Age", "Fare", "Sex","Embarked","Pclass","Pizdatyi"]].corr(), cmap="YlGnBu")
```
