```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
data = pd.read_csv('./9249290-cost-revenue-clean.csv')
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel("Production budget, $")
plt.xlabel("Worldwide gross, $")
plt.ylim(0,3e9)
plt.xlim(0,0.45e9)
plt.show()
```

```python
X1 = pd.DataFrame([1,2,3,4,5]) #створюємо датафрейм для навчання
y1 = pd.DataFrame([100,200,300,400,500]) #створюємо відповідність на осі y для X1
X2 = pd.DataFrame([6,7,8,9,10]) #створюємо датафрейм для тестування
reg = LinearRegression() #обєкт регресії
reg.fit(X1,y1) #навчання моделі
reg.predict(X2) #перевірка
```
