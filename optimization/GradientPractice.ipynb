```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2 + x + 1
def df(x):
    return 2*x+1
```

```python
x_1 = np.linspace(-500,500,200)
```

```python
#Gradient
new_x = -500
prev_x = 0
stepMilti = 0.05
precision = 0.1

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(1000):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x-stepMilti*gradient
    
    step_size = abs(new_x-prev_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if (step_size < precision):
        print("Loop run many times:", n)
        break
```

```python
#plot
plt.figure(figsize=[25,8])
plt.subplot(1,2,1)
plt.scatter(x_1, f(x_1), color="g")
plt.scatter(x_list, f(np.array(x_list)), color="r")
plt.subplot(1,2,2)
plt.scatter(x_1, df(x_1), color="r")
```
