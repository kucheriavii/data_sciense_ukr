# Notebooks Imports and Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 - A simple cost function

## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2+x+1
```

```python
#Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
#plot
plt.xlim(-5,5)
plt.ylim(0,15)
plt.plot(x_1, f(x_1))
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.show()
```

# Slope and Derivatives

```python
def df(x):
    return 2*x+1
```

```python
#plot function and derivative side by side
plt.figure(figsize=[15,5])
# 1 Chart: cost function
plt.subplot(1,2,1)

plt.title('Cost function', fontsize=17)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.plot(x_1, f(x_1), color='b', linewidth=3)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

#2 Chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of the cost function', fontsize=17)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()
plt.show()
```

```python
#plot function and derivative side by side
plt.figure(figsize=[5,17])
# 1 Chart: cost function
plt.subplot(2,1,1)

plt.title('Cost function', fontsize=17)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.plot(x_1, f(x_1), color='b', linewidth=3)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

#2 Chart: Derivative
plt.subplot(2,1,2)
plt.title('Slope of the cost function', fontsize=17)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()
plt.show()
```
