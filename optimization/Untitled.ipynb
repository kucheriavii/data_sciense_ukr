```python
import numpy as np
import matplotlib.pyplot as plt

# Згенеруємо випадкові дані
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Додаємо стовпець одиниць до X для зсуву
X_b = np.c_[np.ones((100, 1)), X]

# Ініціалізуємо параметри
theta = np.random.randn(2, 1)

# Встановлюємо крок навчання
learning_rate = 0.01
n_iterations = 1000

for iteration in range(n_iterations):
    # Обчислюємо градієнт
    gradients = 1/100 * X_b.T.dot(X_b.dot(theta) - y)
    
    # Оновлюємо параметри
    theta = theta - learning_rate * gradients

# Отримуємо оптимальні ваги
optimal_weights = theta

# Візуалізація результатів
plt.scatter(X, y, alpha=0.5, label='Дані')
plt.plot(X, X_b.dot(optimal_weights), color='red', label='Прогнозована лінія')
plt.xlabel('Площа будинку')
plt.ylabel('Ціна будинку')
plt.legend()
plt.show()
```
